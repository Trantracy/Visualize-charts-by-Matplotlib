## Excercise 2: Get Total profit of all months and show line plot with the following Style properties

Generated line plot must include following Style properties: –

* Line Style dotted and Line-color should be red
* Show legend at the lower right location.
* X label name = Month Number
* Y label name = Sold units number
* Add a circle marker.
* Line marker color as read
* Line width should be 3




```
profitList = df ['total_profit'].tolist()
monthList  = df ['month_number'].tolist()

fig, ax = plt.subplots()
plt.xticks(monthList)
ax.plot(monthList,profitList,"ro", linestyle='dashed',markerfacecolor='k', label = " Profit data of last year", linewidth =3)
ax.set(xlabel='Month Number', ylabel='Sold units number',
       title='Company Sales data of last year')
ax.grid(b=None)
plt.legend(loc='lower right')
plt.show()
```

## Exercise 3: Read all product sales data and show it  using a multiline plot

Display the number of units sold per month for each product using multiline plots. (i.e., Separate Plotline for each product for each product).

```
monthList  = df ['month_number'].tolist()
faceCremSalesData   = df ['facecream'].tolist()
faceWashSalesData   = df ['facewash'].tolist()
toothPasteSalesData = df ['toothpaste'].tolist()
bathingsoapSalesData   = df ['bathingsoap'].tolist()
shampooSalesData   = df ['shampoo'].tolist()
moisturizerSalesData = df ['moisturizer'].tolist()

fig, ax = plt.subplots()
plt.xticks(monthList)
plt.plot(monthList, faceCremSalesData, "ro", label = 'Face Cream Sales Data', linestyle='solid', linewidth =3)
plt.plot(monthList, faceWashSalesData, 'go', label = 'face Wash Sales Data', linestyle='solid', linewidth =3)
plt.plot(monthList, toothPasteSalesData, 'bo', label = 'tooth Paste Sales Data', linestyle='solid',  linewidth =3)
plt.plot(monthList, bathingsoapSalesData,'yo', label = 'bathing soap Sales Data' , linestyle='solid', linewidth =3)
plt.plot(monthList, shampooSalesData, 'ko', label = 'shampoo Sales Data' , linestyle='solid', linewidth =3)
plt.plot(monthList, moisturizerSalesData, 'mo', label = 'moisturizer Sales Data', linestyle='solid',linewidth =3)

ax.set(xlabel='Month Number', ylabel='Sales units in number',
       title='Sales data')
ax.grid(b=None)
ax.legend()
plt.show()

```

## Exercise 5: Read face cream and facewash product sales data and show it using the bar chart

Bar chart should display the number of units sold per month for each product. Add a separate bar for each product in the same chart.

```
monthList = df ['month_number'].tolist()
faceCremSalesData   = df ['facecream'].tolist()
faceWashSalesData   = df ['facewash'].tolist()

X = np.arange(1,len(monthList)+1)

plt.bar(X, faceCremSalesData, color = 'b', label='Face Cream Sales Data', width = 0.25)
plt.bar(X + 0.25, faceWashSalesData, color = 'g',label='face Wash Sales Data', width = 0.25)
plt.xticks(monthList)
plt.legend()
plt.xlabel('Month Number')
plt.ylabel('Sales units in number')
plt.legend(loc='upper left')
plt.title(' Facewash and facecream sales data')
plt.show()
```

## Exercise 9: Read Bathing soap facewash of all months and display it using the Subplot

## Exercise 10: Read all product sales data and show it using the stack plot

```

x = df['month_number'].tolist()
y1 = df['facecream'].tolist()
y2 = df['facewash'].tolist()
y3 = df['shampoo'].tolist()
y4 = df['moisturizer'].tolist()
y5 = df['bathingsoap'].tolist()
y6 = df['toothpaste'].tolist()

y = np.vstack([y1, y2, y3, y4, y5, y6])
labels = 'facecream', 'facewash', 'toothpaste', 'bathingsoap','shampoo','moisturizer'

fig, ax = plt.subplots()
ax.stackplot(x, y1, y2, y3, y4, y5, y6, labels=labels)
ax.legend(loc='upper left')

plt.xlabel('Month number')
plt.ylabel('Sales units in number')
plt.title('All product sales dara using stack plot')
plt.show()

```
